### 1.1. Preliminaries: Nernst potential, reversal potential, and resting potential

### 1.1.1. Nernst potential

***TL;DR:*** When at thermal equilibrium, fewer ions are in a state of high energy and more at low energy, which results in a potential difference.

From the theory of thermodynamics, we have:

\begin{align*}
p(E)\propto \exp\left(-\frac{E}{kT}\right),
\end{align*}

where $p(E)$ is the probability of a particle having energy $E$, $k$ is the Boltzmann constant, and $T$ is the temperature.

For an ion with charge $q$ at location $x$ in a static electrical field, the energy is given by: 

\begin{align*}
E = qu(x),
\end{align*}

where $u(x)$ is the potential at $x$.

Interpreting the probability distribution $p(E)$ in terms of the concentration $n(x)$ of ions, we have:

\begin{align*}
n(x)\propto \exp\left(-\frac{qu(x)}{kT}\right).
\end{align*}

For ions at two locations $x_1$ and $x_2$, we have:

\begin{align*}
\frac{n(x_1)}{n(x_2)} = \exp\left(-\frac{q[u(x_1)-u(x_2)]}{kT}\right).
\end{align*}
Thus, the potential difference $\Delta u = u(x_1) - u(x_2)$ is

\begin{align*}
\Delta u = -\frac{kT}{q}\ln\frac{n(x_1)}{n(x_2)},
\end{align*}

which is known as the Nernst potential.

### 1.1.2. Reversal potential

***TL;DR:*** Reversal potential is an alias for Nernst potential when the ion concentration difference is created by the ion pumps in the cell membrane.

In the context of neurons, the two locations $x_1$ and $x_2$ can be interpreted as the inside and outside (or reverse) of the cell membrane, respectively. *Ion pumps* in the cell membrane *actively* maintain different ion concentrations ($n(x_1)$ and $n(x_2)$), which in turn create a potential difference across the membrane, according to the Nernst equation.

In addition to ion pumps, the interior of the cell and the surrounding liquid are in contact through *ion channels*, which *passively* allow ions to move across the membrane. Taking sodium as an example. If we artificially fix the potential difference $\Delta u$ smaller than $E_\mathrm{Na}$, the sodium ions will flow into the cell through the ion channels to reduce the concentration difference. Conversely, if we set $\Delta u$ larger than $E_\mathrm{Na}$, the sodium ions will flow out of the cell to increase the concentration difference. Finally, when $E_\mathrm{Na}$ achieves $\Delta u$, there is no ion flow, and new equilibrium is established. Therefore, the ion flow direction is reversed when the potential difference $\Delta u$ passes $E_\mathrm{Na}$, and that's why we name the Nernst potential as reversal potential.

From example above we can see that while Nernst and reversal potential are defined and calculated at equilibrium state, we can still use them as reference points to understand the ion flow dynamics when the neuron is not at equilibrium.

### 1.2. Membrane potential dynamics of HH model

The standard HH model can be formulated as

$$C\frac{\mathrm dV}{\mathrm dt}=-I_{\mathrm{Na}}-I_{\mathrm{K}}-I_{\mathrm{L}}+I_{\mathrm{syn}}+I_{\mathrm{app}},$$

where

\begin{align*}
I_{\mathrm{Na}}&=g_{\mathrm{Na}}m^3h(V-E_{\mathrm{Na}}),\\
I_{\mathrm{K}}&=g_{\mathrm{K}}n^4(V-E_{\mathrm{K}}),\\
I_{\mathrm{L}}&=g_{\mathrm{L}}(V-E_{\mathrm{L}}).
\end{align*}

Notation is listed as follows:
|Notation|Description|Dimension|
|:---|:---|:---|
|$C$|membrane capacitance|$\mu F/cm^2$|
|$V$|membrane potential|$mV$|
|$I_{\mathrm{Na}}$|sodium current|$\mu A/cm^2$|
|$I_{\mathrm{K}}$|potassium current|$\mu A/cm^2$|
|$I_{\mathrm{L}}$|leak current|$\mu A/cm^2$|
|$I_{\mathrm{syn}}$|synaptic input current|$\mu A/cm^2$|
|$I_{\mathrm{inj}}$|injected input current|$\mu A/cm^2$|
|$g_{\mathrm{Na}}$|maximal sodium conductance|$mS/cm^2$|
|$g_{\mathrm{K}}$|maximal potassium conductance|$mS/cm^2$|
|$g_{\mathrm{L}}$|leak conductance|$mS/cm^2$|
|$E_{\mathrm{Na}}$|sodium reversal potential|$mV$|
|$E_{\mathrm{K}}$|potassium reversal potential|$mV$|
|$E_{\mathrm{L}}$|leak reversal potential|$mV$|
|$m$|sodium activation gating variable|
|$h$|sodium inactivation gating variable|
|$n$|potassium activation gating variable|

## 2. RTM model

In RTM model, only variable $n$ obeys first-order kinetics:

$$\frac{\mathrm dn}{\mathrm dt}=\alpha_n(V)(1-n) - \beta_n(V).$$

Variables $m$ is assumed fast and substituted by its steady-state function:

$$m_{\infty}(V)=\frac{\alpha_m(V)}{\alpha_m(V)+\beta_m(V)}.$$

Variable $h$ is set as a linear function of $n$:

$$h=\max(1 - 1.25 n, 0)$$

We use the following parameters for RTM model, which are originally from [1].

Parameters:

- $C = 1\mu F/\mathrm{cm}^2$
---
- $g_{\mathrm{Na}}=100 \mathrm{mS/cm}^2$
- $E_{\mathrm{Na}}=50 \mathrm{mV}$
- $\alpha_m(V)= \frac{0.32(V+54)}{1-\exp(-0.25(V+54))}$
- $\beta_m(V)=\frac{0.28(V+27)}{\exp[0.2(V+27)-1]}$
---
- $g_{\mathrm{K}}=80 \mathrm{mS/cm}^2$
- $E_{\mathrm{K}}=-100\mathrm{mV}$
- $\alpha_n(V)=\frac{0.032(V+52)}{1-\exp(-0.2(V+52))}$
- $\beta_n(V) =0.5\exp\left(-0.025(V+57)\right)$
---
- $g_{\mathrm{L}} = 0.1\mathrm{mS/cm}^2$
- $E_{\mathrm{L}} = -67\mathrm{mV}$

## 3. WB model

In WB model, variables $h,n$ follow first-order kinetics:

$$
\frac{\mathrm dx}{\mathrm dt}=\phi\left[\alpha_x(V)(1-x)-\beta_x(V)x\right],\quad x\in\{h,n\},
$$

where $\phi$ is a scaling factor. Variable $m$ is also assumed fast and substituted by its steady-state function 

$$m_{\infty}(V)=\frac{\alpha_m(V)}{\alpha_m(V)+\beta_m(V)}.$$

We use the following parameters for WB model, which are originally from [2].

Parameters:

- $C = 1\mu F/\mathrm{cm}^2$
---
- $g_{\mathrm{Na}}=35 \mathrm{mS/cm}^2$
- $E_{\mathrm{Na}}=55 \mathrm{mV}$
- $\alpha_m(V)= \frac{0.1(V+35)}{1-\exp(-0.1(V+35))}$
- $\beta_m(V)=4\exp\left(\frac{-(V+60)}{18}\right)$
- $\alpha_h(V)=0.07\exp\left(-0.05(V+58)\right)$
- $\beta_h(V)=\frac{1}{\exp(-0.1(V+28)) + 1}$
---
- $g_{\mathrm{K}}=9 \mathrm{mS/cm}^2$
- $E_{\mathrm{K}}=-90 \mathrm{mV}$
- $\alpha_n(V)= \frac{0.01(V+34)}{1-\exp(-0.1(V+34))}$
- $\beta_n(V) = 0.125\exp\left(-0.0125(V+44)\right)$
---
- $g_{\mathrm{L}} = 0.1 \mathrm{mS/cm}^2$
- $E_{\mathrm{L}} = -65 \mathrm{mV}$
---
- $\phi = 5$

> Note that synaptic weights and network connectivity are not included here. They will be introduced in the following sections.

### 1.4 Implementation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# =========================================================
# Base Neuron Model Class
# =========================================================
class NeuronModel:
    def __init__(self, params):
        self.params = params
        
    def alpha_beta(self, V):
        raise NotImplementedError
        
    def derivatives(self, state, t, I_inj):
        raise NotImplementedError

# =========================================================
# RTM model
# Corresponds to pyramidal excitatory neurons
# 2 state variables [V, n], h is a function of n
# =========================================================
class RTMNeuron(NeuronModel):
    def __init__(self):
        params = {
            'C': 1.0,
            'g_Na': 100.0, 'E_Na': 50.0,
            'g_K': 80.0,   'E_K': -100.0,
            'g_L': 0.1,    'E_L': -67.0
        }
        super().__init__(params)

    def derivatives(self, state, t, I_inj):
        V, n = state
        p = self.params
        eps = 1e-8

        if callable(I_inj):
            I_t = I_inj(t)
        else:
            I_t = I_inj
        
        # alpha_m, beta_m
        # alpha_m = 0.32(V+54) / (1 - exp(-0.25(V+54)))
        # beta_m = 0.28(V+27) / (exp(0.2(V+27)) - 1)
        num_am = 0.32 * (V + 54)
        den_am = 1.0 - np.exp(-0.25 * (V + 54))
        alpha_m = num_am / (den_am + eps) if abs(den_am) > eps else 0.32 / 0.25
        num_bm = 0.28 * (V + 27)
        den_bm = np.exp(0.2 * (V + 27)) - 1.0
        beta_m = num_bm / (den_bm + eps) if abs(den_bm) > eps else 0.28 / 0.2
        m_inf = alpha_m / (alpha_m + beta_m)

        # alpha_n, beta_n
        # alpha_n = 0.032(V+52) / (1 - exp(-0.2(V+52)))
        # beta_n = 0.5exp(-0.025(V+57))
        num_an = 0.032 * (V + 52)
        den_an = 1.0 - np.exp(-0.2 * (V + 52))
        alpha_n = num_an / (den_an + eps) if abs(den_an) > eps else 0.032 / 0.2 # L'Hopital rule
        beta_n = 0.5 * np.exp(-(V + 57) / 40.0)
        
        # h is a function of n
        h = max(1.0 - 1.25 * n, 0.0)
        
        # channel currents
        I_Na = p['g_Na'] * (m_inf**3) * h * (V - p['E_Na'])
        I_K  = p['g_K']  * (n**4) * (V - p['E_K'])
        I_L  = p['g_L']  * (V - p['E_L'])
        
        # update states
        dVdt = (-I_Na - I_K - I_L + I_t) / p['C']
        dndt = alpha_n * (1 - n) - beta_n * n
        
        return [dVdt, dndt]

# =========================================================
# WB model
# Corresponds to PV+ inhibitory neurons
# 3 state variables [V, h, n]
# =========================================================
class WBNeuron(NeuronModel):
    def __init__(self):
        params = {
            'C': 1.0,
            'g_Na': 35.0,  'E_Na': 55.0,
            'g_K': 9.0,    'E_K': -90.0,
            'g_L': 0.1,    'E_L': -65.0,
            'phi': 5.0
        }
        super().__init__(params)

    def derivatives(self, state, t, I_inj):
        V, h, n = state
        p = self.params
        phi = p['phi']
        eps = 1e-8

        if callable(I_inj):
            I_t = I_inj(t)
        else:
            I_t = I_inj
        
        # alpha_m, beta_m
        # alpha_m = 0.1(V+35) / (1 - exp(-0.1(V+35)))
        # beta_m = 4exp(-(V+60)/18)
        num_am = 0.1 * (V + 35)
        den_am = 1.0 - np.exp(-0.1 * (V + 35))
        alpha_m = num_am / (den_am + eps) if abs(den_am) > eps else 1.0
        beta_m = 4.0 * np.exp(-(V + 60) / 18.0)
        m_inf = alpha_m / (alpha_m + beta_m)
        
        # alpha_n, beta_n
        # alpha_n = 0.01(V+34) / (1 - exp(-0.1(V+34)))
        # beta_n = 0.125exp(-(V+44)/80)
        num_an = 0.01 * (V + 34)
        den_an = 1.0 - np.exp(-0.1 * (V + 34))
        alpha_n = num_an / (den_an + eps) if abs(den_an) > eps else 0.1
        beta_n = 0.125 * np.exp(-(V + 44) / 80.0)

        # alpha_h, beta_h
        alpha_h = 0.07 * np.exp(-(V + 58) / 20.0)
        beta_h  = 1.0 / (np.exp(-0.1 * (V + 28)) + 1.0)
        
        # channel currents
        I_Na = p['g_Na'] * (m_inf**3) * h * (V - p['E_Na'])
        I_K  = p['g_K']  * (n**4) * (V - p['E_K'])
        I_L  = p['g_L']  * (V - p['E_L'])
        
        # update states
        dVdt = (-I_Na - I_K - I_L + I_t) / p['C']
        dhdt = phi * (alpha_h * (1 - h) - beta_h * h)
        dndt = phi * (alpha_n * (1 - n) - beta_n * n)
        
        return [dVdt, dhdt, dndt]
```

```python
# =========================================================
# simulation: inject current to both neurons
# =========================================================
def inject_current(I_rtm, I_wb, t_simulation, dt=0.01, return_trace=False):
    t = np.arange(0, t_simulation, dt)  # seconds
    
    rtm = RTMNeuron()
    wb = WBNeuron()
    
    # initialize states
    state0_rtm = [-67.0, 0.0] # [V, n]
    state0_wb = [-65.0, 0.0, 0.0] # [V, h, n]

    # solve ODEs
    solution_rtm = odeint(rtm.derivatives, state0_rtm, t, args=(I_rtm,))
    solution_wb = odeint(wb.derivatives, state0_wb, t, args=(I_wb,))

    # generate current traces
    if callable(I_rtm):
        I_rtm = [I_rtm(ti) for ti in t]
    else:
        I_rtm = np.full_like(t, I_rtm)
        
    if callable(I_wb):
        I_wb = [I_wb(ti) for ti in t]
    else:
        I_wb = np.full_like(t, I_wb)


    
    # plot results
    plt.figure(figsize=(12, 8))
    
    # plot RTM
    plt.subplot(3, 1, 1)
    plt.title(f"RTM neuron, no synaptic input")
    plt.plot(t, solution_rtm[:, 0], color='red', alpha=0.7)
    plt.ylabel("V (mV)")
    plt.grid(True, which='both', linestyle='--', alpha=0.5)
    # plt.legend()
    
    # plot WB
    plt.subplot(3, 1, 2)
    plt.title(f"WB neuron, no synaptic input")
    plt.plot(t, solution_wb[:, 0], color='blue', alpha=0.7)
    plt.ylabel("V (mV)")
    plt.xlabel("Time (ms)")
    plt.grid(True, which='both', linestyle='--', alpha=0.5)

    # plot injected currents
    plt.subplot(3, 1, 3)
    plt.title("Input current")
    plt.plot(t, I_rtm, 'r', label='RTM Input', alpha=0.7)   # type: ignore
    plt.plot(t, I_wb, 'b', label='WB Input', alpha=0.7) # type: ignore
    plt.ylabel("$I$ ($\mu A/cm^2$)")
    plt.legend(loc="upper right")
    plt.grid(True, linestyle='--', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    if return_trace:
        return solution_rtm[:, 0], solution_wb[:, 0], I_rtm, I_wb
    else:
        return
```

```python
# play with the models :)
inject_current(I_rtm=1.0, I_wb=1.0, t_simulation=100.0, dt=0.01)
```

## 4. Frequency analysis

### 4.1 Fourier analysis of sub-threshold dynamics.

```python
T = 1000
dt = 0.01
amplitude = 0.1
offset = 0.05
f_start = 0
f_end = 100
```

```python
def make_zap_current(t_start, t_end, amplitude, offset, f_start, f_end):
    delta_freq = (f_end - f_start) / ((t_end - t_start) / 1000.0)   # Hz/s
    def zap_func(t):
        if t_start <= t <= t_end:
            t_sec = (t - t_start) / 1000.0
            # Phi = 2 * pi * (f_start * t_sec + 0.5 * k * t_sec^2), 
            # which is the integral of f(t) = f_start + k*t
            phase = 2 * np.pi * (f_start * t_sec + 0.5 * delta_freq * (t_sec**2))
            return amplitude * np.sin(phase) + offset
        else:
            return 0.0
    return zap_func

zap_func = make_zap_current(t_start=0, t_end=T, amplitude=amplitude, offset=offset, f_start=f_start, f_end=f_end)
mem_rtm, mem_wb, input, _ = inject_current(
    I_rtm=zap_func, I_wb=zap_func, t_simulation=T, dt=dt, return_trace=True)   # type: ignore
```

```python
from scipy.fft import rfft, rfftfreq

def analyze_fft_freq(mem_rtm, mem_wb, input, t_simulation, dt, f_start, f_end):
    mem_rtm_fft = rfft(mem_rtm - np.mean(mem_rtm))
    mem_wb_fft = rfft(mem_wb - np.mean(mem_wb))
    input_fft = rfft(input - np.mean(input))
    freqs = rfftfreq(len(mem_rtm), dt / 1000)   # Hz
    
    # compute magnitude impedance
    eps = 1e-8
    Z_rtm = np.abs(mem_rtm_fft) / (np.abs(input_fft) + eps)
    Z_wb = np.abs(mem_wb_fft) / (np.abs(input_fft) + eps)
    mask = (freqs >= f_start) & (freqs <= f_end)

    # plot results
    plt.figure(figsize=(14, 10))
    t = np.arange(0, t_simulation, dt)
    
    # input current
    plt.subplot(2, 2, 1)
    plt.title(f"Input ZAP Current ({f_start} - {f_end} Hz)")
    plt.plot(t, input, 'k', alpha=0.5)
    plt.ylabel(r"$I_{\text{input}}$ ($\mu A/cm^2$)")
    plt.grid(True, linestyle='--', alpha=0.3)
    
    # impedance profile
    plt.subplot(2, 2, 2)
    plt.title("Impedance Profile ($Z(f) = |V(f)/I(f)|$)")
    plt.plot(freqs[mask], Z_rtm[mask], 'r', label='RTM (Excitatory)', alpha=0.7)
    plt.plot(freqs[mask], Z_wb[mask], 'b', label='WB (Inhibitory)', alpha=0.7)
    plt.ylabel(r"Impedance ($k\Omega \cdot cm^2$)")
    plt.xlabel("Frequency (Hz)")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.3)
    
    # RTM voltage trace
    plt.subplot(2, 2, 3)
    plt.title("RTM Voltage Response")
    plt.plot(t, mem_rtm, 'r', alpha=0.7)
    plt.ylabel("V (mV)")
    plt.grid(True, linestyle='--', alpha=0.3)

    # WB voltage response
    plt.subplot(2, 2, 4)
    plt.title("WB Voltage Response")
    plt.plot(t, mem_wb, 'b', alpha=0.7)
    plt.ylabel("V (mV)")
    plt.xlabel("Time (ms)")
    plt.grid(True, linestyle='--', alpha=0.3)

    plt.tight_layout()
    plt.show()

analyze_fft_freq(mem_rtm, mem_wb, input, t_simulation=T, dt=dt, f_start=f_start, f_end=f_end)
```
