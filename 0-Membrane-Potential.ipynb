# 膜电位

任意一个系统处于热平衡时，其中粒子处在能级$E$的概率服从玻尔兹曼分布：

\begin{align*}
p(E)\propto \exp\left(-\frac{E}{kT}\right),
\end{align*}

其中，$p(E)$是粒子处在能级$E$的概率，$k$是玻尔兹曼常数，$T$是温度。

下面从最大化熵的角度简单（不太完备地）证明这一结论。

<div style="background-color: #f1f7ff; border-left: 8px solid #67696dff; padding: 18px; border-radius: 4px; color: #333;">

<b style="color: #67696dff; font-size: 1.2em; display: block; margin-bottom: 10px;">玻尔兹曼分布的推导</b>

考虑一个具有 $N$ 个状态的系统，每个状态 $i$ 具有能量 $E_i$（$N$ 和 $E_i$ 为常数）。一个粒子可以占据这些状态中的任意一个。令 $p_i$ 为粒子占据状态 $i$ 的概率。在热平衡时，系统的自由能最小，当存在下述两个约束条件时，自由能最小等价于熵最大：

- 总概率之和为 1

- 总能量守恒

下面根据最大化熵的条件推导玻尔兹曼分布。熵 $S$ 由下式给出：

$$S = -k \sum_{i=1}^{N} p_i \ln p_i,$$

其中 $k$ 是玻尔兹曼常数。

现在将优化问题形式化如下：

$$\mathrm{maximize\space} S, \mathrm{s.t.} \sum_{i=1}^{N} p_i = 1 \text{ and } \sum_{i=1}^{N} p_i E_i = U.$$

引入拉格朗日乘子来求解该带约束的优化问题：

$$\mathcal{L}(p_1, p_2, \ldots, p_N, \alpha, \beta) = -k \sum_{i=1}^{N} p_i \ln p_i - \alpha \left(\sum_{i=1}^{N} p_i - 1\right) - \beta \left(\sum_{i=1}^{N} p_i E_i - U\right).$$

对每个 $p_i$ 求 $\mathcal{L}$ 的导数得

$$\frac{\partial \mathcal{L}}{\partial p_j} = -k (\ln p_j + 1) - \alpha - \beta E_j.$$

令导数为零可得

$$p_j = \exp\left(-1 - \frac{\alpha}{k}\right)\cdot\exp\left(- \frac{\beta E_j}{k}\right).$$

利用约束条件 $\sum_{i=1}^{N} p_i = 1$ 得到 $\alpha$ 与 $\beta$ 的关系：

$$\exp\left(-1 - \frac{\alpha}{k}\right) = \frac{1}{\sum_{i=1}^{N} \exp\left(- \frac{\beta E_i}{k}\right)}.$$

因此 $p_j = \frac{\exp\left(- \frac{\beta E_j}{k}\right)}{Z}$，其中 $Z = \sum_{i=1}^{N} \exp\left(- \frac{\beta E_i}{k}\right)$。

下面求解$\beta$，将 $p_j$ 代回 $S$，我们有

$$S = -k\sum_{i=1}^N p_i\left(-\frac{\beta E_i}{k} - \ln Z\right) = k\ln Z + \beta U.$$

计算 $S$ 的全微分：

\begin{align*} \mathrm dS &= \mathrm d(k\ln Z + \beta U) \\ 
&= \frac{k}{Z} \cdot \mathrm dZ + \beta \mathrm dU + U \mathrm d\beta \\
 &= \frac{k}{Z} \cdot \left[ \sum_{i=1}^N \left(-\frac{E_i}{k}\right) \exp\left(- \frac{\beta E_i}{k}\right)\mathrm d\beta \right] + \beta \mathrm dU + U\mathrm d\beta \\
  &= \frac{-\sum_{i=1}^N E_i \exp\left(- \frac{\beta E_i}{k}\right) \mathrm d\beta}{Z} + \beta \mathrm dU + U\mathrm d\beta \\ 
&= \frac{-(\sum_{i=1}^N E_i p_i Z) \mathrm d\beta}{Z} + \beta \mathrm dU + U\mathrm d\beta \\
 &= -U \mathrm d\beta + \beta \mathrm dU + U\mathrm d\beta \\
  &= \beta\mathrm dU. \end{align*}

因此，$\frac{\mathrm dS}{\mathrm dU} = \beta$。

假设系统体积 $V$ 恒定，根据热力学第一定律，我们有 $dU = TdS$，即$\frac{\mathrm dS}{\mathrm dU} = \frac{1}{T}$。

因此，$\beta = \frac{1}{T}$，从而有

$$p(E) = \frac{\exp\left(- \frac{E}{kT}\right)}{\sum_{i=1}^{N} \exp\left(- \frac{E_i}{kT}\right)},$$

这正是玻尔兹曼分布，证毕。

在静电场中，带电粒子的能级$E$由电荷量和电势给出：

\begin{align*} E = qu(x), \end{align*}

其中 $u(x)$ 是 $x$ 处的电势，$q$ 是粒子的电荷量。

当粒子数量很多时，可以将热平衡时粒子的概率分布 $p(E)$ 看作粒子浓度 $n(x)$：

\begin{align*} n(x)\propto \exp\left(-\frac{qu(x)}{kT}\right). \end{align*}

对于位于两个位置 $x_1$ 和 $x_2$ 处的电荷量为$q$的粒子，我们有：

\begin{align*} \frac{n(x_1)}{n(x_2)} = \exp\left(-\frac{q[u(x_1)-u(x_2)]}{kT}\right). \end{align*}

因此，电势差 $\Delta u = u(x_1) - u(x_2)$ 为

\begin{align*} \Delta u = -\frac{kT}{q}\ln\frac{n(x_1)}{n(x_2)}, \end{align*}

该电势差被称为能斯特电位。

## 2. Ion Pumps and Channels

***TL;DR:*** Ion pumps consume energy to maintain ion concentration gradients, while ion channels use the gradients to create electric potentials by passively transporting ions across membrane.

### 2.1. Ion pumps

In the practice above, we can observe that there are much higher concentrations of K⁺ inside the cell and Na⁺ outside the cell. So why and how do cells collect K⁺ while evicting Na⁺?

The reason for such phenomenon may be attributed to evolution. One of the hypotheses is that life originated in a K⁺-rich environment (unlike today's ocean, which is rich in Na⁺), making it easy for cells to retain K⁺ and keep them inside the membrane. And K⁺ is utilized by many cellular organelles and processes, making it essential for life. As time goes by, environment changed and Na⁺ became more and more while K⁺ got less (especially in ocean). To Adapt to such changes, cells evolved specialized proteins called Na⁺-K⁺ pump that **actively** imports K⁺ and exports Na⁺.

Here we take the concentration gradients of Na⁺ and K⁺ as given facts due to the existence of Na⁺-K⁺ pump, and focus on how potential difference across the membrane is generated and maintained in the following sections.

### 2.2. Ion channels

Neurons also have specialized proteins called ion channels embedded in their membranes that allow specific ions to pass through **without consuming energy** (e.g., ATP). These channels can be selective for certain ions (e.g., Na⁺, K⁺, Ca²⁺, Cl⁻) and can open or close in response to various stimuli, such as changes in voltage (voltage-gated channels), binding of a ligand (ligand-gated channels), or mechanical forces (mechanically-gated channels).

As we will see in the next section, ion channels play a crucial role in generating the membrane potential by allowing ions to move across the membrane according to their concentration gradients established by the ion pumps.

### 2.3. Generation of potential difference across membrane

Let's start from an idealized neuron. There is only K⁺ inside and outside the cell and there is no Na⁺. K⁺ is more inside the cell and can move across the membrane through K⁺ channels freely.

At the beginning, K⁺ ions tend to diffuse from inside to outside the cell due to the concentration gradient. As K⁺ ions leave the cell, *the inside of the cell becomes negatively charged while the outside becomes positively charged*. This charge separation creates an electric field across the membrane that opposes further diffusion of K⁺ ions.

When the electric force pushing K⁺ ions back into the cell balances the diffusion process, the system reaches equilibrium, and the potential difference will be exactly the Nernst potential for K⁺ ions calculated above.
